#### Autocorrelation Analysis

**Moran's I** is a measure of sparial correlation, which is characterized by a correlation in a signal among nearby loations in space.

The core idea of a local Moran’s I is to identify cases in which the value of an observation and the average of its surroundings is either more similar or dissimilar than we would expect from pure chance. The mechanism to do this is similar to global Moran’s I which yields only one statistic to summarize the whole study area, but it is applied in this case to each observation.

This analysis will healp us identify hot and cold spots for clustering.

For further reading on Moran's I access [Wikipedia Moran's I](https://en.wikipedia.org/wiki/Moran%27s_I)

```python
y=df_final[2021].values
# compute contiguity weights
w = Queen.from_dataframe(df_final)
# Row-standardization
w.transform = "R"
```

We plot a Moran Local scatterplot which does not differ much from a regular scatterplot. Here we plot pur variable - income 2021 - with its spatial lag. <br><br>
The spatial lag is simply th result of a matrix containg information about which location are considered neighbours and the strength of their spatial connection.

```python
# calculate Moran_Local and plot
moran_loc = Moran_Local(y, w)
fig, ax = moran_scatterplot(moran_loc)
ax.set_xlabel('Income')
ax.set_ylabel('Spatial Lag of Income')
plt.show()
```

There ae 4 quadrants in the scatterplot that represent either similarities or dissimilarities between a specific polygon with its neighbouring polygons. From that, we have a few conclusions that we can take:

- Positive Spatial Autocorrelation: If the Moran's I scatterplot shows a positive trend, where high values of Moran's I are clustered together and low values are clustered together, it suggests that there is positive spatial autocorrelation in your data. In other words, similar values tend to be close to each other, indicating spatial clustering.

- Negative Spatial Autocorrelation: Conversely, if the Moran's I scatterplot shows a negative trend, where high values of Moran's I are surrounded by low values and vice versa, it suggests negative spatial autocorrelation. This means that dissimilar values tend to be close to each other, indicating a spatial dispersion or segregation pattern.

- No Spatial Autocorrelation: If the Moran's I scatterplot shows no discernible pattern, with values scattered randomly across the plot, it suggests that there is no spatial autocorrelation in the data. In this case, the data points are distributed randomly across space.

To statistically compute these autocorrelatios, we use <ins>LISA</ins>, or Local Indicators of Spatial Association.


```python
fig, ax = moran_scatterplot(moran_loc, p=0.05)
ax.set_xlabel('Income')
ax.set_ylabel('Spatial Lag of Income')
plt.show()
```

```python
lisa_cluster(moran_loc, df_final, p=0.05, figsize = (20,20))

plt.show()
```

```python
#plot everything onto a single plot
plot_local_autocorrelation(moran_loc, df_final, 2021, cmap= "PRGn")
plt.show()
```
