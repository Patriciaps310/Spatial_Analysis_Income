```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
from keplergl import KeplerGl
```

#### Data Cleaning

```python
pd.set_option('display.max_rows', None)
```

Exporting data

```python
#export data
exports = pd.read_excel('data\exports_data.xlsx', sheet_name='Sheet 1', skiprows=9)
```

```python
exports
```

```python
#last 3 rows are not necessary
exports_trimmed = exports[:-3]
```

```python
#rename TIME column as country
exports_renamed = exports_trimmed.rename(columns={'TIME': 'country'})

# for i in exports_renamed.columns[1:]:
#     exports_renamed.rename(columns={i: 'exp_'+i}, inplace=True)


exports_renamed.rename(columns={'2022': 'exp_2022'}, inplace=True)
```

```python
exports_2022 = exports_renamed.loc[:,['country','exp_2022']]
```

```python
#drop nan from df
exports_dropna = exports_2022.dropna()
```

```python
#remove columns with : value
for i in exports_dropna.columns[1:]:

    exports_filtered = exports_dropna[~exports_dropna[i].str.contains(':', na=False)]
```

```python
exports_filtered
```

Importing Data

```python
#export data
imports = pd.read_excel('data\imports_data.xlsx', sheet_name='Sheet 1', skiprows=9)
```

```python
imports
```

```python
#last 3 rows are not necessary
imports_trimmed = imports[:-3]
```

```python
#rename TIME column as country
imports_renamed = imports_trimmed.rename(columns={'TIME': 'country'})
# for i in imports_renamed.columns[1:]:
#     imports_renamed.rename(columns={i: 'imp_'+i}, inplace=True)
    
imports_renamed.rename(columns={'2022': 'imp_2022'}, inplace=True)
```

```python
exports_2022 = imports_renamed.loc[:,['country','imp_2022']]
```

```python
exports_2022
```

```python
#drop nan from df
imports_dropna = exports_2022.dropna()
```

```python
#remove columns with : value
for i in imports_dropna.columns[1:]:
    imports_filtered = imports_dropna[~imports_dropna[i].str.contains(':', na=False)]
```

```python
imports_filtered
```

World Geojson 

```python
#import world countries outline
countries_outline=gpd.read_file('data/countries.geojson')
```

```python
countries_outline
```

```python
map_exports =  countries_outline.merge(exports_filtered, left_on = 'ADMIN', right_on='country').drop(['country'], axis=1)
```

```python
map_full =  map_exports.merge(imports_filtered, left_on = 'ADMIN', right_on='country').drop(['country'], axis=1)
```

```python
map_full
```

```python
map1=KeplerGl(height=500,data={'EU Imports ':map_full})
```
